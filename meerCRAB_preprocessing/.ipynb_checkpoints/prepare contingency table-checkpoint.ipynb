```python
import pandas as pd
```

```python
network = 'NET3'
directory1 = '../meerCRAB_output/'+network+'/T9_NRD/classified_examples/'
```

```python
network_output = pd.read_csv(directory1+'overall_probability.csv')
network_output
```

```python
ML_LABEL = pd.DataFrame()
for i in range(network_output.shape[0]):
    if (network_output['ML_PROB_REAL'].iloc[i] >= 0.5):
        label = pd.DataFrame([1],columns=[network+'_ML_LABEL'])
    else:
        label = pd.DataFrame([0],columns=[network+'_ML_LABEL'])
    ML_LABEL = ML_LABEL.append([label],ignore_index=True)
    
ML_LABEL
```

```python
predicton_table = pd.concat([network_output,ML_LABEL],axis=1,join='outer')
predicton_table
```
